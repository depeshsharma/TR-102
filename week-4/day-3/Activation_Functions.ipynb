```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-8, 8, 100)
relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
```

```python
plt.plot(x, relu, label='ReLU')
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.legend()
plt.title("Activation Functions")
plt.grid(True)
```
